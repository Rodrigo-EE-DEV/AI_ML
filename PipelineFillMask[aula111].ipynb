```python
import transformers
from transformers import pipeline
```

```python
mascarar = pipeline("fill-mask", model = "neuralmind/bert-base-portuguese-cased")
texto = mascarar("Existe uma chance do corpo cair no [MASK]")
for x in range(len(texto)):
  print(texto[x])
```

```python
texto2 = mascarar(["Existe uma chance do corpo cair no [MASK]","Brasilia eh a capital do [MASK]"])
for counter in range(len(texto2)):
  print(texto2[counter])
```
