```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
```

```python
spam_obj = pd.read_csv("spam.csv")
spam_obj.head(6)
```

```python
spam_obj.shape
```

```python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(spam_obj["Category"])
print(y)
```

```python
mensagens = spam_obj["Message"].values
x_treino, x_teste, y_treino, y_teste = train_test_split(mensagens, y, test_size = 0.3)
```

```python
token = Tokenizer(num_words = 1000)
token.fit_on_texts(x_treino)
x_treino = token.texts_to_sequences(x_treino)
x_teste = token.texts_to_sequences(x_teste)
```

```python
print(x_treino)
```

```python
x_treino = pad_sequences(x_treino, padding = 'post', maxlen = 500)
x_teste = pad_sequences(x_teste, padding = 'post', maxlen = 500)
```

```python
x_treino
```

```python
token.word_index
```

```python
len(token.word_index)
```

```python
modelo = Sequential()
modelo.add(Embedding(input_dim=len(token.word_index), output_dim=50, input_length=500))
modelo.add(Flatten())
modelo.add(Dense(units = 10, activation = "relu"))
modelo.add(Dropout(0.1))
modelo.add(Dense(units = 1, activation = "sigmoid"))
```

```python
modelo.compile(loss = "mean_squared_error", optimizer = "adam", metrics = ["accuracy"])
```

```python
modelo.summary()
```

```python
modelo.fit(x_treino, y_treino, epochs = 20, batch_size = 10, verbose = True, validation_data = (x_teste, y_teste))
```

```python
loss, accuracy = modelo.evaluate(x_teste, y_teste)
print("Loss: ", loss)
print("Accuracia: ", accuracy)
```

```python
nova_previsao = modelo.predict(x_teste)
print(nova_previsao)
```

```python
prev = nova_previsao>0.5
print(prev)
```

```python
cm = confusion_matrix(y_teste, prev)
print(cm)
```
