```python
iris = datasets.load_iris()
clust_obj1 = scale(iris.data)
clust_obj2 = scale(np.random.rand(150,4))
```

```python
print(clust_obj2)
```

```python
inertia  = []
for i in range(1,8):
  kmeans = KMeans(n_clusters = i, n_init = "auto")
  kmeans.fit(clust_obj1)
  inertia.append(kmeans.inertia_)
plt.plot(range(1,8), inertia)
plt.title("Elbow")
plt.grid(True)
plt.show()
```
