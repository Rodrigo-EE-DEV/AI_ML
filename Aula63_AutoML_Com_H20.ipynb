```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
data_inst = h2o.import_file("insurance.csv")
```

```python
data_inst = data_inst.drop("C1")
```

```python
treino, teste = data_inst.split_frame(ratios = [0.7], seed = 0)
```

```python
model_automl_obj = H2OAutoML(max_runtime_secs = 180, sort_metric = "AUTO") ##Nao use AUC, a classificaçao nao binaria, portanto, nao serve para esse caso.
model_automl_obj.train(y = "Accident", training_frame = treino)
```

```python
ranking = model_automl_obj.leaderboard
ranking = ranking.as_data_frame()
ranking
```

```python
prever = model_automl_obj.leader.predict(teste)
```
